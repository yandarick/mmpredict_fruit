## 划分训练集和验证集

https://mmpretrain.readthedocs.io/zh_CN/latest/user_guides/dataset_prepare.html


````python
import os
import shutil
import random

# 数据集所在的根目录
root_dir = r'/content/drive/MyDrive/OpenMMlab/MMPretrain/mmpretrain/data/fruit30'

# 划分比例
train_ratio = 0.8

# 训练集和验证集目录, 不需要事先建立
train_dir = r'/content/drive/MyDrive/OpenMMlab/MMPretrain/mmpretrain/data/fruit/train'
val_dir = r'/content/drive/MyDrive/OpenMMlab/MMPretrain/mmpretrain/data/fruit/val'

# 遍历所有文件夹
for folder in os.listdir(root_dir):
    folder_path = os.path.join(root_dir, folder)
    if os.path.isdir(folder_path):
        # 获取当前文件夹中所有图片的路径
        image_paths = [os.path.join(folder_path, filename) for filename in os.listdir(folder_path) if filename.endswith('.jpg')]
        print(image_paths)
        # 随机打乱图片路径
        random.shuffle(image_paths)
        print(image_paths)
        # 计算训练集和测试集的分界点
        print(len(image_paths))
        split_index = int(len(image_paths) * train_ratio)
        print(split_index)
        # 将图片复制到训练集和测试集目录下
        for i, image_path in enumerate(image_paths):
            if i < split_index:
                dst_dir = os.path.join(train_dir, folder)
            else:
                dst_dir = os.path.join(val_dir, folder)
            if not os.path.exists(dst_dir):
                os.makedirs(dst_dir)
            shutil.copy(image_path, dst_dir)
````
